#**SEABORN**


Seaborn library is based on matplotlib. Seaborn is built on top of matplotlib and integrates closely with pandas data structures, making it an excellent tool for visualizing data from DataFrames and arrays. It also provides a variety of built-in themes and color palettes to customize visualizations.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

### Why Seaborn?

- provides a layer of abstraction hence simpler to use
- better aesthetics
- more graphs included

### **Seaborn Roadmap**

Types of Functions : there are two levels of functions for creating plots

- Figure Level (Seaborn recommends using figure-level functions whenever possible because they provide a higher-level interface that simplifies the creation of complex multi-plot figures. Figure-level functions like *sns.relplot()* and *sns.catplot()* handle many details automatically)
- Axis Level

Main Classification

- Relational Plot (scatter plot, line plot)
- Distribution Plot (histogram, Kde plot, rugplot)
- Categorical Plot (bar, count plot, box plot, violin plot, swam plot)
- Regression Plot
- Matrix Plot (heatmeap, clustermap)
- Multiplots (join plot, pair plot)

https://seaborn.pydata.org/api.html

#**1. Relational Plot**

- to see the statistical relation between 2 or more variables.
- Bivariate Analysis

**Plots under this section**

- scatterplot
- lineplot

**name of figure level function:**  relplot()

**name of axes level function:**  scatterplot(), lineplot()

```python
tips = sns.load_dataset('tips')
tips
```

## Scatter Plot

```python
# scatter plot --> axes level function

sns.scatterplot(data=tips, x='total_bill', y='tip',
                hue='sex', style='time', size='size',
)
```

```python
# scatter plot --> figure level function

sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter',
            hue='sex', style='time', size='size', height=4, aspect=1.5,
)
```

## Line Plot

```python
gap = px.data.gapminder()
gap
```

```python
temp_df = gap[gap['country'] == 'India']
temp_df
```

```python
# line plot --> axes level function

sns.lineplot(data=temp_df, x='year', y='lifeExp',)
```

```python
# line plot --> figure level function

sns.relplot(data=temp_df, x='year', y='lifeExp', kind='line',height=4, aspect=1.5, )
```

```python
new_temp_df = gap[gap['country'].isin(['India', 'Brazil', 'Germany'])]
new_temp_df.sample(5)
```

```python
# hue in Seaborn allows you to visually represent an additional categorical variable
# in your plot by coloring the data points accordingly.

sns.relplot(kind='line', data=new_temp_df, x='year', y='lifeExp',
            hue='country', style='continent', size='continent',
            height=4, aspect=1.5,
)
```

## facet plot

```python
# facet plot work only with figure level function (means: relplot)
# output: <seaborn.axisgrid.FacetGrid at 0x793c3d22a620>
# axes level function can't do this
# aspect ratio = 1.5, means width is 1.5 times the height

sns.relplot(data= tips, x='total_bill', y='tip', kind='scatter',
        col='sex', height=3, aspect=1.5,
)
```

```python
sns.relplot(data= tips, x='total_bill', y='tip', kind='scatter',
        col='smoker', row='sex',
        height=3, aspect=1.5,
)
```

## Wrap column (col_wrap= no. of graph in one column)

```python
new_gap = px.data.gapminder()
new_gap.sample(5)
```

```python
# show graph between lifeExp and gdp for all years
sns.relplot(data=new_gap , x='lifeExp', y='gdpPercap', kind='scatter',
            height=4, aspect=1.5, col='year', col_wrap=4,
)
```

# **2. Distribution Plots**

- used for univariate analysis (focus on one column)
- used to find out the distribution
- Range of the observation (what is the range of distribution?)
- Central Tendency (what is mean/median of data?)
- is the data bimodal? (are there multiple peaks in data?)
- Are there outliers?

*Plots under distribution plot:*

- histplot
- kdeplot
- rugplot

**name of figure level function:**  displot()

**name of axes level function:**  histplot(), kdeplot(), rugplot()

#histplot

```python
tips = sns.load_dataset('tips')
tips
```

```python
# axes level function
sns.histplot(data=tips, x='total_bill')
```

```python
# axes level function
sns.histplot(data=tips, x='total_bill', bins=20)
```

```python
# figure level function
sns.displot(data=tips, x='total_bill', kind='hist')
```

```python
# Itâ€™s also possible to visualize the distribution of a categorical variable using the logic of a histogram.
# Discrete bins are automatically set for categorical variables

# we can call it 'countplot'
sns.displot(data=tips, x='day', kind='hist')
```

```python
tips.sample(5)
```

```python
# hue parameter
sns.displot(data=tips, x='tip', kind='hist', hue='sex')
```

```python
# hue parameter and element='step'

sns.displot(data=tips, x='tip', kind='hist',
            hue='sex', element='step',
)
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
sns.displot(data=titanic, x='age', kind='hist',
            hue='sex', element='step',
)
```

```python
# faceting only works with figure level function

tips = sns.load_dataset('tips')
sns.displot(data=tips, x='tip', kind='hist',
            col='sex', element='step',
)
```

#kdeplot

Rather than using discrete bins, a KDE plot smooths the observations with a Gaussian kernel, producing a continuous density estimate

```python
sns.kdeplot(data=tips, x='total_bill')
```

```python
# figure level function
sns.displot(data=tips, x='total_bill', kind='kde',
            hue= 'sex', fill=True, height=4, aspect=2,
)
```

# rugplot

Plot marginal distributions by drawing ticks along the x and y axes.

This function is intended to complement other plots by showing the location of individual observations in an unobtrusive way.

```python
sns.kdeplot(data=tips, x='total_bill')
sns.rugplot(data=tips, x='total_bill')
```

### Bivariate histogram

A bivariate histogram bins the data within rectangles that tile the plot and then shows the count of observations within each rectangle with the fill color


```python
# sns.histplot(data=tips, x='total_bill', y='tip')
sns.displot(data=tips, x='total_bill', y='tip', kind='hist')

# reading graph: the combination of '15 usd total bill and 3 usd tip' is maximum in our data
```

### Bivariate Kdeplot

a bivariate KDE plot smoothes the (x, y) observations with a 2D Gaussian


```python
sns.kdeplot(data=tips, x='total_bill', y='tip')
```

#**3. Matrix Plot**

- Heatmap
- Clustermap

```python
new_gap = px.data.gapminder()
new_gap.sample(5)
```

#Heatmap

```python
# Plot rectangular data as a color-encoded matrix
temp_gap_df = new_gap.pivot(index='country',columns='year',values='lifeExp')
temp_gap_df.sample(5)
```

```python
# axes level function, no figure level function

plt.figure(figsize=(10,10))
sns.heatmap(temp_gap_df)
```

```python
# graph only for european contries

eur_gap_df = new_gap[new_gap['continent'] == 'Europe'].pivot(index='country',columns='year',values='lifeExp')
eur_gap_df.sample(6)
```

```python
# Annotation, cmap, linewidth

plt.figure(figsize=(10,10))
sns.heatmap(eur_gap_df, annot=True, linewidth=0.5, cmap='viridis')
```

#Clustermap

Plot a matrix dataset as a hierarchically-clustered heatmap.
This function requires scipy to be available.


```python
iris = px.data.iris()
iris
```

```python
plt.figure(figsize=(5,10))
sns.clustermap(iris.iloc[:,[0,1,2,3]])  # send all rows and column 0,1,2,3
```
